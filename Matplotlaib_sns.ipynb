```python
figg = plt.figure(facecolor='grey',edgecolor='green')
axis = figg.add_axes([1,1,1,1])
axis.plot(x,y)



axis2 = figg.add_axes([1.3,1,.6,.5])      #means that we access the canvas below the plot
axis2.plot(x,y)
```
