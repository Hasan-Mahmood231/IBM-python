```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/TelecomCustomerChurn.csv')
```

```python

df['MonthlyCharges_bin'] = pd.qcut(df['MonthlyCharges'], q=10, duplicates='drop')

# Plot the distribution
sns.countplot(x='MonthlyCharges_bin', data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
df['Tenure'] = pd.qcut(df['Tenure'],q=5)
sns.countplot(x= 'Tenure', data= df)
plt.show()
```
